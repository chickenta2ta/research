```python
import os
import cv2


def video_to_image(path_to_video, path_to_output):
  cap = cv2.VideoCapture(path_to_video)
  
  i = 1
  while True:
    ret, img = cap.read()
    if ret:
      cv2.imwrite(os.path.join(path_to_output, f"{i:d}.jpg"), img)
      i += 1
    else:
      return

video_to_image("/content/drive/My Drive/Colab Notebooks/Research/Raw Data/20221024/14B_1.MOV", "/content/drive/My Drive/Colab Notebooks/Research/Eggplant Tracking/data2/14B")
```

```python
import glob


for path_to_image in glob.glob("/content/drive/My Drive/Colab Notebooks/Research/Eggplant Tracking/data2/14B/*.jpg"):
  image = cv2.imread(path_to_image)
  basename = os.path.basename(path_to_image)
  
  rotated_image = cv2.rotate(image, cv2.ROTATE_180)

  path_to_output = os.path.join("/content/drive/My Drive/Colab Notebooks/Research/Eggplant Tracking/data2/14B", basename)
  cv2.imwrite(path_to_output, rotated_image)
```
