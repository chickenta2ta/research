```python
import json


json_in_file = "annotations.json"
with open(json_in_file) as f_in:
    imgs_anns = json.load(f_in)

    json_dict = {}
    json_dict["images"] = imgs_anns["images"]
    json_dict["categories"] = imgs_anns["categories"]

    ann_dicts = []
    for ann in imgs_anns["annotations"]:
        ann_dict = {}
        ann_dict["num_keypoints"] = ann["num_keypoints"]
        ann_dict["keypoints"] = ann["keypoints"]
        ann_dict["segmentation"] = [
                [ann["segmentation"][0][0],
                 ann["segmentation"][0][1],
                 ann["segmentation"][0][2],
                 ann["segmentation"][0][1],
                 ann["segmentation"][0][2],
                 ann["segmentation"][0][3],
                 ann["segmentation"][0][0],
                 ann["segmentation"][0][3]]
            ]
        ann_dict["iscrowd"] = ann["iscrowd"]
        ann_dict["area"] = ann["area"]
        ann_dict["image_id"] = ann["image_id"]
        ann_dict["bbox"] = ann["bbox"]
        ann_dict["category_id"] = ann["category_id"]
        ann_dict["id"] = ann["id"]
        ann_dicts.append(ann_dict)
    json_dict["annotations"] = ann_dicts

    json_out_file = "annotations_out.json"
    with open(json_out_file, "w") as f_out:
      json.dump(json_dict, f_out, indent=4)
```
